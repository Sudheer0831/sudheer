```python
import pandas as pd
```

```python
df=pd.read_csv("mental_health_workplace_survey.csv")
```

```python
df.isnull().sum() # gives the sum of missing data in dataframe
```

```python
num_cols = df.select_dtypes(include='number').columns  
for col in num_cols:
    median_val = df[col].median()
    df[col] = df[col].fillna(median_val)
    cat_cols = df.select_dtypes(include='object').columns   
for col in cat_cols:
    mode_val = df[col].mode()[0]
    df[col] = df[col].fillna(mode_val)
```

```python
df.head()
```

```python
df['workHoursPerDay'] = df['WorkHoursPerWeek'] / 7    #we are creating a new column-work HoursPerDay
df['workHoursPerDay']
```

```python
df['Worked'] = df['workHoursPerDay'] >8
df['Worked'].value_counts()
```

```python
median_workhours = df['workHoursPerDay'].median()
df['workHoursPerDay'] = df['workHoursPerDay'].apply(lambda x: median_workhours if x > 8 else x)
```
